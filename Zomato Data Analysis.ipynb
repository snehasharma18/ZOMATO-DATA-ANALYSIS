# FILLING OF NULL VALUES

Analyzing Cost column

```python
data['Cost'].unique()
```

Removing commas

```python
data["Cost"]=data["Cost"].str.replace(",","")
data.Cost.unique()
```

Filling null values using bfill method

```python
data['Cost'].fillna(method='ffill', inplace=True)
```

Converting the datatype of cost from object to int

```python
data["Cost"] = data["Cost"].astype(int)
```

Checking datatype

```python
data.Cost.info()
```

Analyzing rest_type column

```python
data.rest_type.unique()
```

Count of unique values

```python
len(data.rest_type.unique())
```

There are some duplicate rest_type values like 'Cafe, Casual Dining' and 'Casual Dining, Cafe' or 'Quick Bites, Cafe' and 'Cafe, Quick Bites', We clean these duplications as it occurs only to the resturants that have 2 rest_type categories combined

```python
data['rest_type'] = data['rest_type'].apply(lambda x: ', '.join(sorted(x.split(', '))) if pd.notna(x) else x)
```

Checking for any duplicates

```python
data['rest_type'].unique()
```

Checking for the restaurants which have null rest_type value

```python
null_rest_type=data.loc[data['rest_type'].isnull(),'Restaurant'].unique()
null_rest_type
```

Filling null values using mode

```python
data.rest_type.fillna(data.rest_type.mode()[0],inplace=True)
```

Checking if the null values are filled

```python
data.rest_type.isnull().sum()
```

Analyzing dish_liked column

```python
data["dish_liked"].unique()
```

Identifying the most occuring dish liked by the customers

```python
data["dish_liked"].value_counts()
```

Filling the null values with mode

```python
data.dish_liked.fillna(data.dish_liked.mode()[0],inplace=True)
```

Checking if all null values are filled

```python
data.dish_liked.isnull().sum()
```

Analyzing Rate column

```python
data.rate.unique()
```

To remove leading and tailing white spaces from rate column

```python
data["rate"]=data["rate"].str.strip()
data.rate.unique()
```

Remove -

```python
data["rate"]=data["rate"].str.replace("-","")
data.rate.unique()
```

Remove "NEW"

```python
data["rate"]=data["rate"].str.replace("NEW","")
data.rate.unique()
```

Remove "/5"

```python
data["rate"]=data["rate"].str.replace("/5","")
data.rate.unique()
```

```python
data.rate.info()
```

The pd.to_numeric() function is used to convert the "rate" column to numeric values, and the errors="coerce" parameter ensures that any non-convertible values (including empty strings) are replaced with NaN. After this process, the "rate" column should only contain numeric values or NaN.

```python
data["rate"] = pd.to_numeric(data["rate"], errors="coerce")
```

```python
data.rate.isnull().sum()
```

```python
data.rate.isnull().sum()/len(data)*100
```

```python
data.rate.describe()
```

```python
null_rate_rows = data[data['rate'].isnull()][['Restaurant', 'rate']]
null_rate_rows
```

Out of a total of 51717 Restaurants, 10052 of them have not been rated.

So we fill the null values using the average rating.

```python
data.rate.fillna(data.rate.mean(),inplace=True)
```

```python
data.rate.isnull().sum()
```

top 6 locations having over 2.5K restaurants

```python
loc_index=data.groupby('City')['Restaurant'].size().sort_values(ascending=False).iloc[0:6].index
f,a=plt.subplots(2,3,figsize=(15,10),dpi=100)
for i,axes_subplot in enumerate(a.flatten()):
    sns.histplot(data=data,x=data.loc[data["City"]==loc_index[i],'rate'],bins=6,ax=axes_subplot)
    axes_subplot.set_title(f"{i+1} {loc_index[i]}")
plt.show()
```

```python
# accept online orders, accept booking table, accept both, don't accept both
accept_only_online_orders = data.online_order.value_counts().loc['Yes'] - data[(data['online_order']=='Yes') & (data['book_table']=='Yes')].book_table.value_counts().loc['Yes']
#accept booking table
accept_only_booking_table = data.book_table.value_counts().loc['Yes']-data[(data['online_order']=='Yes') & (data['book_table']=='Yes')].book_table.value_counts().loc['Yes']
#accept both
accept_both = data[(data['online_order']=='Yes') & (data['book_table']=='Yes')].book_table.value_counts().loc['Yes']
#don't accept both
dont_accept_both = data[(data['online_order']=='No') & (data['book_table']=='No')].book_table.value_counts().loc['No']
```

Pie Chart For Restaurant that accept online orders and booking tables or not

```python
size=[accept_only_online_orders,accept_only_booking_table,accept_both,dont_accept_both]
labels=['Accept only Online orders','Accept only Booking table','Accept Both',"Don't Accept both"]
colors=['gold','orange','coral','olive']
explode=(.01,.01,.01,.01)
plt.figure(figsize=(4,4),dpi=100)
plt.pie(size,autopct='%.1f%%',colors=colors,labels=labels,explode=explode)
plt.axis('equal')
plt.show()
```
